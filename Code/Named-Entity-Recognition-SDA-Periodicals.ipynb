```python
import nltk
import os
import pandas as pd
```

```python
source_dir = "../Data/Periodical-text-files-single-pages/"

texts = []

for filename in os.listdir(source_dir):
    with open(os.path.join(source_dir, filename), 'r') as obit:
        content = obit.read()
    texts.append(
        {
            "doc_id": filename,
            "text": content
        }
    )
    
texts_df = pd.DataFrame(texts)

texts_df.head()
```

```python
texts_df.iloc[3]['text']

LibM040426 = texts_df.iloc[3]['text']
```

```python
LibM040426
```

```python
l_sentences = LibM040426.split('.')
```

```python
l_sentences
```

```python
l_parsed = []

for sentence in l_sentences:
    tokens = nltk.word_tokenize(sentence)
    l_parsed.append(tokens)
    
l_parsed
```

```python
l_tagged = []

for parsed_sentence in l_parsed:
    tagged = nltk.pos_tag(parsed_sentence)
    l_tagged.append(tagged)
```

```python
l_tagged
```

```python
l_entities = []

for tagged_sent in l_tagged:
    entities = nltk.chunk.ne_chunk(tagged_sent)
    l_entities.append(entities)
```

```python
l_entities
```
